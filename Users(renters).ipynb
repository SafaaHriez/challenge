<h2>This notebook presents the data from the perspective of the users </h2>

```python
import pandas as pd
import numpy as np
import seaborn as sns  
import matplotlib                 
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(8,8)})
```

```python
data = pd.read_csv('Airbnb.csv')
```

<h3> 1. The first figure</h3>

```python
sns.catplot('room_type',kind = 'count',data = data)
print('Figure 1: density of each room type')
```

Figure 1 illustrates the density of each room type. As you can see in the figure, the most listings are of type home or apartments, followed by the private room type. It is easily to find one of these room types but for shared type there is small number of listings in this type so you may face difficulties finding listings of this type.

<h3> 2. The second figure</h3>

```python
#500 is used to eliminate the outliers and make the figure clearer
df2=data[data.price < 400]
df2 = df2.sort_values('room_type',ascending=True)
fig3=sns.violinplot(data=df2, x='room_type', y='price')
fig3.set_title('Figure 2: density and distribution of prices for each room type')
```

Figure 2 shows the density and distributions of price for each room type. It is clear that the highest median price is for home or apartment with 150 dollars per night. The distribution of the price for homes or apartments is between 50 to 350 dollars, between 20 to 200 dollars for private rooms and between 5 to 130 dollars for shared room. The density of the price for home/apartment is evenly distributed between 75 and 200. But for the private and shared rooms, the most dense price is around the median, 75 dollars for private room, 40 dollars for shared room.   

<h3> 3. The third figure </h3>

```python

print('Figure 3: density and distribution of prices for each room type based on the neighberhood group')

f, subplots = plt.subplots(len(data.neighbourhood_group.unique()), 1, figsize=(10, 20))

for i, neighbourhood_group in enumerate(data.neighbourhood_group.unique()):
    neighbourhoods = data.loc[data.neighbourhood_group == neighbourhood_group]
    neighbourhoods = neighbourhoods.sort_values('room_type',ascending=True)
    ax = subplots[i]
    df4=neighbourhoods[neighbourhoods.price < 400 ]
    df4=df4[df4.price > 5]
    
    sns.violinplot(data=df4, x='room_type', y='price', ax=ax)
    ax.axhline(0, color="k", clip_on=False)
    ax.set_title(neighbourhood_group)



sns.despine(bottom=True)
plt.tight_layout(h_pad=2)
plt.show()

```

Figure 3 presents the density and distribution of prices for each room type based on the neighberhood group.

  <li><b>For Manhattan:</b></li>
  The distribution of the price for the entire home or apartment is between 50 and 400 dollars. For private room, the price is between 10 and 250 dollar. For shared room, the price is between 5 and 120 dollars.
